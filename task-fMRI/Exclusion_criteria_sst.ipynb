#### 5. beta weights outliers: hemispheric mean beta weights within 2 s.d. of the sample mean

```python
atlas_mapping_df = pd.read_csv(os.path.join(datasets_path, "atlas mapping", "Destrieux_Yeo17_mapping_table.csv"))
atlas_mapping_df.head()
```

```python
left_hemi_columns = [item for item in atlas_mapping_df["Destrieux atlas label"].tolist() if item.startswith("L")]
sst_activate_df["left_hemi_mean_act_beta"] = 0
left_hemi_vertices_sum = 0
for parcel in left_hemi_columns:
    parcel_vertices_num = atlas_mapping_df.loc[atlas_mapping_df["Destrieux atlas label"] == parcel, "vertices_Count"].iloc[0]
    left_hemi_vertices_sum += parcel_vertices_num
    sst_activate_df["left_hemi_mean_act_beta"] += parcel_vertices_num*sst_activate_df[parcel]
sst_activate_df["left_hemi_mean_act_beta"]/=left_hemi_vertices_sum
print(sst_activate_df["left_hemi_mean_act_beta"].describe())
sst_activate_df.head()
```

```python
right_hemi_columns = [item for item in atlas_mapping_df["Destrieux atlas label"].tolist() if item.startswith("R")]
sst_activate_df["right_hemi_mean_act_beta"] = 0
right_hemi_vertices_sum = 0
for parcel in right_hemi_columns:
    parcel_vertices_num = atlas_mapping_df.loc[atlas_mapping_df["Destrieux atlas label"] == parcel, "vertices_Count"].iloc[0]
    right_hemi_vertices_sum += parcel_vertices_num
    sst_activate_df["right_hemi_mean_act_beta"] += parcel_vertices_num*sst_activate_df[parcel]
sst_activate_df["right_hemi_mean_act_beta"]/=right_hemi_vertices_sum
print(sst_activate_df["right_hemi_mean_act_beta"].describe())
sst_activate_df.head()
```

```python
left_hemi_mean_beta = sst_activate_df["left_hemi_mean_act_beta"].mean()
left_hemi_std_beta = sst_activate_df["left_hemi_mean_act_beta"].std()

left_hemi_upper_bound = left_hemi_mean_beta + 2 * left_hemi_std_beta
left_hemi_lower_bound = left_hemi_mean_beta - 2 * left_hemi_std_beta

right_hemi_mean_beta = sst_activate_df["right_hemi_mean_act_beta"].mean()
right_hemi_std_beta = sst_activate_df["right_hemi_mean_act_beta"].std()

right_hemi_upper_bound = right_hemi_mean_beta + 2 * right_hemi_std_beta
right_hemi_lower_bound = right_hemi_mean_beta - 2 * right_hemi_std_beta
```

```python
sst_activate_df = sst_activate_df[
    (sst_activate_df["left_hemi_mean_act_beta"] >= left_hemi_lower_bound) &
    (sst_activate_df["left_hemi_mean_act_beta"] <= left_hemi_upper_bound)
]
print("Total subjects(after s5, left):", len(sst_activate_df["src_subject_id"].value_counts()))
print("total scans(after s5, left):", len(sst_activate_df))
sst_activate_df = sst_activate_df[
    (sst_activate_df["right_hemi_mean_act_beta"] >= right_hemi_lower_bound) &
    (sst_activate_df["right_hemi_mean_act_beta"] <= right_hemi_upper_bound)
]
print("Total subjects(after s5, right):", len(sst_activate_df["src_subject_id"].value_counts()))
print("total scans(after s5, right):", len(sst_activate_df))
```

#### 6. behavioral QC: meet task-specific performance criteria

Poor performance in the SST task (mr_y_tfmri__sst__beh__qc_indicator = 0) is indicated if:

1. Go trials
+ <300 GO trials(2 runs)
+ Correct GO trial percentage <60%
+ Incorrect GO trial percentage >30%
+ Late GO trial percentage (summing correct and incorrect trials) >30%
+ No Response GO trial percentage >30%
+ Correct GO response time < Incorrect Stop response time

2. STOP trials
+ STOP trial accuracy (i.e., successful inhibitions) <20% or >80%

```python
sst_beh_qc_df = pd.read_csv(os.path.join(datasets_path, "qc", "mri_y_tfmr_sst_beh.csv"))
sst_beh_qc_df["tfmri_sst_beh_performflag"].head()
```

```python
sst_beh_qc_df = sst_beh_qc_df[sst_beh_qc_df["tfmri_sst_beh_performflag"] == 1]

keys_from_qc = sst_beh_qc_df[['src_subject_id', 'eventname']]
sst_activate_df = pd.merge(
    sst_activate_df,
    keys_from_qc,
    on=['src_subject_id', 'eventname'],
    how='inner'
)
print("Total subjects(after s6):", len(sst_activate_df["src_subject_id"].value_counts()))
print("total scans(after s6):", len(sst_activate_df))
```
