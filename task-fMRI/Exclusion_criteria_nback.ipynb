#### 5. beta weights outliers: hemispheric mean beta weights within 2 s.d. of the sample mean

```python
atlas_mapping_df = pd.read_csv(os.path.join(datasets_path, "atlas mapping", "Destrieux_Yeo17_mapping_table.csv"))
atlas_mapping_df.head()
```

```python
left_hemi_columns = [item for item in atlas_mapping_df["Destrieux atlas label"].tolist() if item.startswith("L")]
nback_activate_df["left_hemi_mean_act_beta"] = 0
left_hemi_vertices_sum = 0
for parcel in left_hemi_columns:
    parcel_vertices_num = atlas_mapping_df.loc[atlas_mapping_df["Destrieux atlas label"] == parcel, "vertices_Count"].iloc[0]
    left_hemi_vertices_sum += parcel_vertices_num
    nback_activate_df["left_hemi_mean_act_beta"] += parcel_vertices_num*nback_activate_df[parcel]
nback_activate_df["left_hemi_mean_act_beta"]/=left_hemi_vertices_sum
print(nback_activate_df["left_hemi_mean_act_beta"].describe())
nback_activate_df.head()
```

```python
right_hemi_columns = [item for item in atlas_mapping_df["Destrieux atlas label"].tolist() if item.startswith("R")]
nback_activate_df["right_hemi_mean_act_beta"] = 0
right_hemi_vertices_sum = 0
for parcel in right_hemi_columns:
    parcel_vertices_num = atlas_mapping_df.loc[atlas_mapping_df["Destrieux atlas label"] == parcel, "vertices_Count"].iloc[0]
    right_hemi_vertices_sum += parcel_vertices_num
    nback_activate_df["right_hemi_mean_act_beta"] += parcel_vertices_num*nback_activate_df[parcel]
nback_activate_df["right_hemi_mean_act_beta"]/=right_hemi_vertices_sum
print(nback_activate_df["right_hemi_mean_act_beta"].describe())
nback_activate_df.head()
```

```python
left_hemi_mean_beta = nback_activate_df["left_hemi_mean_act_beta"].mean()
left_hemi_std_beta = nback_activate_df["left_hemi_mean_act_beta"].std()

left_hemi_upper_bound = left_hemi_mean_beta + 2 * left_hemi_std_beta
left_hemi_lower_bound = left_hemi_mean_beta - 2 * left_hemi_std_beta

right_hemi_mean_beta = nback_activate_df["right_hemi_mean_act_beta"].mean()
right_hemi_std_beta = nback_activate_df["right_hemi_mean_act_beta"].std()

right_hemi_upper_bound = right_hemi_mean_beta + 2 * right_hemi_std_beta
right_hemi_lower_bound = right_hemi_mean_beta - 2 * right_hemi_std_beta
```

```python
nback_activate_df = nback_activate_df[
    (nback_activate_df["left_hemi_mean_act_beta"] >= left_hemi_lower_bound) &
    (nback_activate_df["left_hemi_mean_act_beta"] <= left_hemi_upper_bound)
]
print("Total subjects(after s5, left):", len(nback_activate_df["src_subject_id"].value_counts()))
print("total scans(after s5, left):", len(nback_activate_df))

nback_activate_df = nback_activate_df[
    (nback_activate_df["right_hemi_mean_act_beta"] >= right_hemi_lower_bound) &
    (nback_activate_df["right_hemi_mean_act_beta"] <= right_hemi_upper_bound)
]
print("Total subjects(after s5, right):", len(nback_activate_df["src_subject_id"].value_counts()))
print("total scans(after s5, right):", len(nback_activate_df))
```
