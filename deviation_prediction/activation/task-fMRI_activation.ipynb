```python
import pandas as pd
import numpy as np
```

```python
activation_table = pd.read_csv('D:/code/SC_ADHD/datasets/ABCD/task-fMRI/activation/SCdata_Yeo17_CV75_Activation.sum.msmtcsd.combat_TD_ADHDall.csv')
# activation_table.head()
print(activation_table.shape)
```

```python
supplement_table = pd.read_csv('D:/code/SC_ADHD/datasets/ABCD/task-fMRI/deviation/SCdeviation_nbackActivation_deviation_ADHD_TDtest.csv', low_memory=False)
print(supplement_table.shape)
supplement_table.head()
```

```python
feature_columns = [
    "SC.109_deviationZ", "SC.111_deviationZ", "SC.116_deviationZ", "SC.118_deviationZ",
    "SC.119_deviationZ", "SC.120_deviationZ", "SC.18_deviationZ", "SC.57_deviationZ",
    "SC.59_deviationZ", "SC.60_deviationZ", "SC.65_deviationZ", "SC.70_deviationZ",
    "SC.79_deviationZ", "SC.84_deviationZ", "SC.8_deviationZ", "SC.93_deviationZ",
    "SC.98_deviationZ"
]
predict_columns = [
    "nbackFP.A_deviationZ",
    "nbackFP.B_deviationZ",
    "nbackSM.A_deviationZ",
    "nbackSM.B_deviationZ",
    "nbackVA.B_deviationZ"
]
covariates_list = ["age", "sex", "if_TD"]

supplement_table = supplement_table[feature_columns + ["scanID", "age", "sex", "if_TD"]]
print(supplement_table.shape)
supplement_table.head()
```

```python
mapping_dict = {
    "Yeo17_3_h": "SM.A_h",
    "Yeo17_4_h": "SM.B_h",
    "Yeo17_8_h": "VA.B_h",
    "Yeo17_12_h": "FP.A_h",
    "Yeo17_13_h": "FP.B_h"
}
FD_columns = ["meanFDnback", "meanFDsst", "meanFDmid"]
left_columns = ["scanID"] + FD_columns

original_columns = activation_table.columns.tolist()
column_mapping = {}

for col in original_columns:
    new_col = col
    for key, value in mapping_dict.items():
        if key in col:
            new_col = col.replace(key, value)
            break
    column_mapping[col] = new_col
activation_table = activation_table.rename(columns=column_mapping)

modified_columns = [new_col for old_col, new_col in column_mapping.items() if old_col != new_col]
columns_to_keep = left_columns + modified_columns
existing_columns = [col for col in columns_to_keep if col in activation_table.columns]

activation_table = activation_table[existing_columns]
print(activation_table.shape)
activation_table.head()
```

```python
demo_table = supplement_table.merge(activation_table, on='scanID', how='left')
print(demo_table.shape)
demo_table.head()
```

```python
demo_table.to_csv("D:/code/SC_ADHD/datasets/ABCD/task-fMRI/activation/demo_table.csv", index=False)
```
